```python
import urllib.request
```

```python
url="https://www.mohfw.gov.in/"
req = urllib.request.Request(url)
resp = urllib.request.urlopen(req)
respData = resp.read()
```

```python
import pandas as pd
```

```python
lists=pd.read_html(respData)
type(lists)
```

```python
df=lists[0]
```

```python
df.drop('S. No.', axis=1, inplace=True)
df
```

```python
df.drop(range(35, 41), axis=0,  inplace=True)
```

```python
df.dtypes
```

```python
df.rename({'Name of State / UT':'state', 'Total Confirmed cases*':'confirmed', 'Cured/Discharged/Migrated':'recovered', 'Deaths**':'deaths'}, axis=1, inplace=True)
df
```

```python
df.dtypes
```

```python
df.confirmed=df.confirmed.map(int)
```

```python
df.recovered=df.recovered.map(int)
df.deaths=df.deaths.map(int)
```

```python
df.dtypes
```

```python
t_confirmed=df.confirmed.sum()
```

```python
t_recovered=df.recovered.sum()
```

```python
t_death=df.deaths.sum()
```

```python
t_rate=t_recovered*100/t_confirmed
t_rate #recovery rate
```

```python
df.sort_values(by='confirmed', inplace=True, ascending=False)
df
```
