```python
from bs4 import BeautifulSoup
```

```python
import urllib.request
url="https://www.mohfw.gov.in/"
req = urllib.request.Request(url)
resp = urllib.request.urlopen(req)
respData = resp.read()
```

```python
s=BeautifulSoup(respData, parser='html5lib') #html5lib, html5, html, html.parser, xml, lxml.. any from thiese parser
```

```python
print(s.get_text(strip=True, separator="\n"))
```

```python
link=s.find("a") #getting links
link
```

```python
link.get_text() #geeting text
```

```python
link.get('href') #getting attribute value
```

```python
s.find_all("a") #getting all anchors
```

```python
s.find_all
```

```python
img_tag=s.find_all("img")
img_tag
```

```python
home="https://www.mohfw.gov.in/"
for x in img_tag:
    src=x.get('src')
    name=src.split("/")[-1]
    file=open("images/"+name, "wb")
    try:
        file.write(urllib.request.urlopen(home+src).read())
    except:
        pass
    file.close()
    
```
