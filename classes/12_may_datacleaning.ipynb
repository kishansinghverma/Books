```python
import pandas as pd
```

```python
df=pd.read_csv('c:/dataset/analysis/weather_data.txt')
df
```

```python
#how to find missing values in all cols
df.isnull().sum()
```

```python
#fill all missing values with same value
df.fillna(10)
```

```python
#fill missing values column wise
df.fillna({'temperature':15,'windspeed':9,'event':'not given'})
```

```python
#fill missing values column wise with mean
df.fillna({'temperature':df.temperature.mean(),'windspeed':df.windspeed.mean(),'event':'not given'})
```

```python
df
```

```python
#filling missing values using backward fill
df.bfill()
```

```python
#filling missing values using forward fill
df.ffill()
```
