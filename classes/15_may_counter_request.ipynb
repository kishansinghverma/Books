```python
import slate3k
```

```python
file=open('c:/dataset/pdf_files/array.pdf','rb')
pages=slate3k.PDF(file)
text=pages.text()
```

```python
from collections import Counter
ctr=Counter(['sonu','monu','sonu','chintu','monu','sonu'])
print(ctr)
```

```python
import re
words=re.findall('[\w]+',text)#returns a list containing words(1 to N length) extracted from text
print(text)
print(words)
```

```python
ctr=Counter(words)
print(ctr)
```

```python
print(ctr.most_common(10))
```

```python
import os
import docxpy
path='c:/dataset/docx_files/'
word=input('enter word to be matched:')
all_file_names=os.listdir(path)#returns a list of all file names present in given dir
print("Total Files:",len(all_file_names))
docx_file_names=[]
for filename in all_file_names:
    if(filename.endswith('.docx')):
        docx_file_names.append(filename)
print("Docx Files:",len(docx_file_names))

result_file_names=[]
for filename in docx_file_names:
    text=docxpy.process(path+filename)
    if(word in text):
        result_file_names.append(filename)
print("Matched Files:",len(result_file_names)) 
for file in result_file_names:
    print(file)
```
