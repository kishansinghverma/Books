```python
import requests
resp=requests.get('https://www.mohfw.gov.in/')
#print(resp)
#print(resp.content) #read webdata as bytes
#print(resp.text) #read webdata as str
web_data=resp.text
```

```python
import pandas as pd
```

```python
#returns list of dataframe objects corresponding to each <table> in webdata
df_list=pd.read_html(web_data)
```

```python
type(df_list)
```

```python
len(df_list)
```

```python
covid19=df_list[0]
```

```python
type(covid19)
```

```python
#delete S.No. column and make changes in existing dataframe
covid19.drop('S. No.',axis=1,inplace=True)
```

```python
#rename columns just for simplicity purpose
covid19.columns=['State','Confirm_cases','Discharged','Deaths']
```

```python
covid19
```

```python
covid19.drop([33,34,35,36,37,38],axis=0,inplace=True)
```

```python
covid19
```

```python
#returns type of values in each column
covid19.dtypes
```

```python
#map(fun) perform operation(fun) on each value of column and returns all values 
covid19.Confirm_cases=covid19.Confirm_cases.map(int)
covid19.Discharged=covid19.Discharged.map(int)
covid19.Deaths=covid19.Deaths.map(int)
covid19.dtypes
```

```python
def death_ratio(death,confirm):
    return (death/confirm)*100
```

```python
covid19['Death_ratio']=death_ratio(covid19.Deaths,covid19.Confirm_cases)
```

```python
covid19[covid19.Death_ratio==covid19.Death_ratio.max()]
```
