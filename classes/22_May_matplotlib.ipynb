```python
plt.figure(figsize=(10,6))
plt.scatter(range(1,7),[40,60,100,150,220,300],c='b')
plt.scatter(range(1,11),[20,25,50,60,67,100,150,160,200,210],c='g')
plt.xticks(range(1,11))
plt.yticks(range(0,301,50))
plt.xlabel('Wickets')
plt.ylabel('Runs')
plt.title("Wickets v/s Runs")
for i in range(1,11,2):
    plt.text(i,200,'Python',fontsize=20)
plt.legend(['Ind','NZ'])
plt.savefig('Ind_vs_Nz.png')
```
