```python
df=pd.read_csv('c:/dataset/analysis/tips.csv')
df
```

```python
sbn.countplot(x='day',data=df)
plt.show()
```

```python
sbn.countplot(x='day',data=df,hue='gender')
plt.show()
```

```python
sbn.countplot(x='day',data=df,hue='time')
plt.show()
```

```python
sbn.jointplot(x='total_bill',y='tip',data=df)#(scatter & histogram)
plt.show()
```

```python
sbn.pairplot(data=df)#(scatter b/w each numeric pair and histogram for each numeric column)
plt.show()
```

```python
sbn.boxplot(x='gender',y='total_bill',data=df)
plt.show()
```
